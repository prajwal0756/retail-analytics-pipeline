```python
import pandas as pd

df = pd.read_csv("../data/raw/online_retail_II.csv", encoding="ISO-8859-1")

df.head()
```

### Here in this dataset some rows of invoice columns contain c in the starting which means the cancellation of the order so a cancelled order cannot give us meaningful insights, so we will remove it from there.

```python
df[df['Invoice'].astype(str).str.startswith('C')].head()
```

```python
df[df['Invoice'].astype(str).str.startswith('C')].iloc[0]
```

```python
df = df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df[df['Quantity'] < 0].head()
```

### Again, there is a negative value in the quantity which I need to remove from the dataset as it doesn't contribute for predictions or analysis.

```python
df.drop(df[df['Quantity'] < 0].index, inplace=True)
```

### Removing Duplicates

```python
df = df.drop_duplicates()
```

```python
df[df['Quantity'] < 0].head()
```

#### Converting InvoiceDate datatype into datetime

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

Introducing new column for better analysis as more features give more ways also it removes unnecessary columns also

```python
df['TotalPrice'] = df['Quantity'] * df['Price']
```
